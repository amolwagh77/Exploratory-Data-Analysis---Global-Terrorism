#### **Missing Values/Null Values**

````python
# Missing Values/Null Values Count

missing_values=data.isnull().sum()
missing_values
````

````python
# Visualizing the missing values

missing_values.plot(kind='bar',color="cornflowerblue",figsize=(15,6),fontsize=13)
plt.xlabel('columns',fontsize=15)
plt.ylabel('Number of missing values',fontsize=15)
plt.title('Missing values present in the dataset',fontsize=17)
plt.show
````

````python
data['Killed']=data['Killed'].fillna(0)
data["Wounded"]=data["Wounded"].fillna(0)
data['Casualty']=data['Killed']+data['Wounded']
````

## **1- YEAR WISE ATTACK**

### **A. Number of attack in each year**

````python
attacks=data["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={'index':'Year','count':'Attack'}).set_index('Year')
attacks.head(5)

````

````python
attacks.plot(kind='bar',color="cornflowerblue",figsize=(15,6),fontsize=13)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Number of attacks',fontsize=15)
plt.title('TOTAL NUMBER OF ATTACKS BETWEEN 1971 TO 2017',fontsize=17)
plt.show
````

### **B. Total casualties in each year (Number of killed + number of wounded )**

````python
Total_casualties=data[['Year','Casualty']].groupby('Year').sum()
Total_casualties.head()
````

````python
Total_casualties.max()
````

````python
Total_casualties.min()
````

````python
Total_casualties.plot(kind='bar',color="cornflowerblue",figsize=(15,6))
plt.title('Total Casualties By Year',fontsize=(14))
plt.xlabel('Year',fontsize=14)
plt.ylabel('Number of casualties',fontsize=14)
````
