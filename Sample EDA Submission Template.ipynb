## **2. REGION WISE ATTACK**

### **(i). Distribution of Terrorist Attack Over a Region** 

````python
Attack_region=pd.crosstab(data.Year,data.Region)
Attack_region.head(10)
````

````python
Attack_region.plot(kind='area',stacked=False,alpha=0.5,figsize=(20,10))
````

### **(ii). Total Attack In Each Region From 1970-2017** 

````python
regt=Attack_region.transpose()
regt['Total']=regt.sum(axis=1)
x=regt['Total'].sort_values(ascending=False)
x
````

````python
x.plot(kind='bar',figsize=(13,3))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
````

### **(iii). Total Casualties In Each Region**

````python
Rc=data[['Region','Casualty']].groupby('Region').sum().sort_values(by='Casualty',ascending=False)
 
````

````python
Rc.plot(kind='bar',color='cornflowerblue',figsize=(15,6))
plt.title("Region wise Casualties",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
````

### **(iv). Killed In Each Region**

````python
rk=data[['Region','Killed']].groupby('Region').sum().sort_values(by='Killed',ascending=False)
rk
````

### **(v). Wounded In Each Region**

````python
rw=data[['Region','Wounded']].groupby('Region').sum().sort_values(by='Wounded',ascending=False)
rw
````

````python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

rk.plot(kind='bar',color='cornflowerblue',figsize=(15,6),ax=ax0)
ax0.set_title('People Killed In Each Region')
ax0.set_xlabel('Regions')
ax0.set_ylabel("Number of people killed")


rw.plot(kind='bar',color='cornflowerblue',figsize=(15,6),ax=ax1)
ax1.set_title('People Wounded In Each Region')
ax1.set_xlabel('Regions')
ax1.set_ylabel('Number of people wounded')

plt.show()
````
